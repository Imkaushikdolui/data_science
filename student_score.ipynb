```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_score.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

### DROP a column

```python
df = df.drop("Unnamed: 0", axis=1)
df.head()
```

### MODIFICATION IN DATA 

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 - 10","5-10")
df.head()
```

### GENDER DISTRIBUTION

```python
plt.figure(figsize=(3,5))
ax = sns.countplot(data= df, x="Gender")
ax.bar_label(ax.containers[0])
plt.title("GENDER DISTRIBUTION")
plt.show()
```

```python
# From the above chart we have analysed that : 
#the number of females are higher then the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(gb, annot= True)
plt.title("RELATIONSHIP BETWEEN PARENT EDUCATION AND STUDENT SCORE")
plt.show()
```

# from the above chart we can conclude that there is impact on the students score due to thier parents education.

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(gb1, annot= True)
plt.title("RELATIONSHIP BETWEEN PARENT MARITIAL STATUS AND STUDENT SCORE")
plt.show()
```

# from the above chart we can conclude that there is negligible impact on students score with relation to the maritial status of thier parents

### Disterbution of Ethnic groups

```python
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count()
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()

l = ["group A","group B","group C","group D","group E",]
plt.pie([groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']],labels = l, autopct= "%1.2f%%")
plt.show()
```
